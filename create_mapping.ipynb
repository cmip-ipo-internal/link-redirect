```python
import pandas as pd 
import re
import numpy as np


def makeint(x):
    try:
        return int(x)
    except:
        return 0
    
```

```python
old = pd.read_xml('oo_old.xml')
```

```python
savecols = ['id', 'current_version',
       'folder_id', 'title', 'content_length',
       'create_by','create_on']
```

```python
old2 = old[savecols][1:].sort_values('current_version')
old2['content_length'] = old2['content_length'].astype(int)
old2['current_version'] = [makeint(i) for i in old2['current_version']]

old2.to_csv('old.csv')
old2.tail()
```

```python
file_content = open('oo_new.txt','r').read()


# columns = ['id', 'version', 'version_group', 'current_version', 'folder_id', 'title', 'content_length', 'file_status', 'category', 'create_by', 'create_on', 'modified_by', 'modified_on', 'tenant_id', 'converted_type', 'comment', 'changes', 'encrypted', 'forcesave', 'thumb']

# Extract values from file_content using regular expression
values = re.findall(r"\((.*?)\)", file_content)



cleaned_values = [re.sub(r'\{[^{}]*\}', '', re.sub(r'\n+', ' ', v.replace('null', '""'))) for v in values[1:]]

# Split values and create a list of tuples
ncoms = 13
# data = [tuple(v.strip("'") for v in val.split(',')) for val in values[1:]]
columns = values[0].replace('`','').split(',')[:ncoms]
data = [v.replace("'",'').replace('"','').replace("'","").split(',')[: ncoms] for v in cleaned_values]

# Create DataFrame
new = pd.DataFrame(data, columns=columns)

columns
```

```python
new2 = new[savecols][1:].sort_values('current_version')


new2['content_length'] = [makeint(i) for i in new2['content_length']]
new2['current_version'] = [makeint(i) for i in new2['current_version']]


new2.to_csv('new.csv')
new2
```
